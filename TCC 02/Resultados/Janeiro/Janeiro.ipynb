```python
import pandas
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
```

```python
df_janeiro = pandas.read_csv("./data/Cluster-Crime-Janeiro.csv")
```

```python
crime_tipos = df_janeiro[['NATUREZA DA OCORRÊNCIA']]
crime_tipo_total = crime_tipos.groupby('NATUREZA DA OCORRÊNCIA').size()
crime_tipo_counts = df_janeiro[['NATUREZA DA OCORRÊNCIA']].groupby('NATUREZA DA OCORRÊNCIA').sum()
crime_tipo_counts['TOTAL'] = crime_tipo_total
all_crime_tipos = crime_tipo_counts.sort_values(by='TOTAL', ascending=False)
```

### Filtro dos 10 crimes com mais ocorrências em janeiro

```python
all_crime_tipos.head(10)
```

```python
all_crime_tipos_top10 = all_crime_tipos.head(10)
all_crime_tipos_top10.plot(kind='barh', figsize=(12,6), color='#3f3fff')
plt.title('Top 10 crimes por tipo (Jan 2017)')
plt.xlabel('Número de crimes')
plt.ylabel('Crime')
plt.tight_layout()
ax = plt.gca()
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
plt.show()
```

### Todas as ocorrências criminais de janeiro

```python
all_crime_tipos
```

### Quantidade de crimes por região

```python
group_df_janeiro = df_janeiro.groupby('CLUSTER')
crimes = group_df_janeiro['NATUREZA DA OCORRÊNCIA'].count()
```

```python
crimes.plot(kind='barh', figsize=(10,7), color='#3f3fff')
plt.title('Número de crimes por região (Jan 2017)')
plt.xlabel('Número')
plt.ylabel('Região')
plt.tight_layout()
ax = plt.gca()
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
plt.show()
```

### As 5 regiões com mais ocorrências

```python
regioes = df_janeiro.groupby('CLUSTER').count()
grupo_de_regioes = regioes.sort_values('NATUREZA DA OCORRÊNCIA', ascending=False)
```

```python
grupo_de_regioes['TOTAL'] = grupo_de_regioes.ID
top_5_regioes_qtd = grupo_de_regioes.TOTAL.head(6)
```

```python
top_5_regioes_qtd.plot(kind='barh', figsize=(10,4), color='#3f3fff')
plt.title('Top 5 regiões com mais crimes')
plt.xlabel('Número de crimes')
plt.ylabel('Região')
plt.tight_layout()
ax = plt.gca()
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
plt.show()
```

### Acima podemos ver que a região 4 teve o maior número de ocorrências criminais

### Podemos agora ver quais são essas ocorrências de forma mais detalhada

```python
regiao_4_detalhe = df_janeiro[df_janeiro['CLUSTER'] == 4]
regiao_4_detalhe
```

### Uma análise sobre as 5 ocorrências mais comuns

```python
crime_types = regiao_4_detalhe[['NATUREZA DA OCORRÊNCIA']]
crime_type_total = crime_types.groupby('NATUREZA DA OCORRÊNCIA').size()
crime_type_counts = regiao_4_detalhe[['NATUREZA DA OCORRÊNCIA']].groupby('NATUREZA DA OCORRÊNCIA').sum()
crime_type_counts['TOTAL'] = crime_type_total
all_crime_types = crime_type_counts.sort_values(by='TOTAL', ascending=False)
```
